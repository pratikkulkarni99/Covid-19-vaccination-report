```python
Vr.columns
#above instruction will show all columns that are present in our dataset
```

```python
sf=Vr.fillna(Vr.mean)
sf=Vr.fillna(0)
sf
#above instructions will fill all na values with mean of repsective columns
```

```python
x=sf[["people_vaccinated","people_fully_vaccinated","daily_vaccinations_raw","daily_vaccinations"]]
y=sf['total_vaccinations']
#x is our independant variables and y is our dependant variables . In this statement we assiqned 4 values for x and 1 value to y variable
```

# Use of Linear regression

```python
#assiqning model to linear regression
model=LinearRegression()
model.fit(x,y)
model.score(x,y)
#use of fit() is use to train a dataset
#use of score is used to get accuracy of our model
```

```python
#it splits the data in two subsets one for train and test
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
ypred=model.predict(xtest)
#assiqning model linear regression and predicting value as per testing
data=pd.DataFrame({"Actual":ytest,"Predicted":ypred})
print(data)
```

```python
print("R square",model.score(xtrain,ytrain))
print("y interpret",model.intercept_)
print("slope",model.coef_)
#getting value of accuracy of linear regression
```

```python
residual=ytest-ypred
residual
#getting value of ytest-ypred
```

```python
np.mean(residual)
#getting mean of residual
```

# To plot a Distplot

# Use of polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_trans=PolynomialFeatures(degree=4)
#importing polynomial regression 
x_poly=poly_trans.fit_transform(x)
#assiqning value of x_poly to poly_trans.fit_transform
```

```python
x_poly
```

```python
poly_model=LinearRegression()
poly_model.fit(x_poly,y)
#assigning polynomial model and fitting x_poly and y values
```

```python
ypred=poly_model.predict(x_poly)
#to get value of ypred 
```

```python
data=pd.DataFrame({"Actual":np.array(y).flatten(),"Predicted":ypred.flatten()})
#to create dataframe of actual and predicted values
```

```python
print(data)
#to print all values of dataset
```

```python
poly_model.score(x_poly,y)
#getting value of accuracy
```

# Use of ridge method

```python
from sklearn.linear_model import Lasso
Ridge_model=Lasso(alpha=0.3)
Ridge_model.fit(x,y)
Ridge_model.score(x,y)
#assiqning ridge_model,fitting model,getting value of accuracy
```
